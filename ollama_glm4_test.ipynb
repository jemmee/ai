```python
import ollama

def local_chat_loop():
    model_name = "glm4"
    
    # Initialize the conversational history with a system behavior rule
    messages = [
        {"role": "system", "content": "You are a concise engineering assistant. Provide technical, punchy answers."}
    ]
    
    print(f"--- Local Chat Initialized using {model_name} (Type 'quit' to exit) ---")
    
    while True:
        user_input = input("\nYou: ")
        if user_input.lower() == "quit":
            break
            
        # Append the new user prompt to the history stack
        messages.append({"role": "user", "content": user_input})
        
        print("AI: ", end="", flush=True)
        
        # Stream the response chunk by chunk to prevent UI lag
        stream = ollama.chat(model=model_name, messages=messages, stream=True)
        
        full_response = ""
        for chunk in stream:
            content = chunk.message.content
            print(content, end="", flush=True)
            full_response += content
            
        # Commit the model's response to memory so follow-up context works
        messages.append({"role": "assistant", "content": full_response})
        print()

if __name__ == "__main__":
    local_chat_loop()
```
